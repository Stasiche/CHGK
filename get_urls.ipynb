```python
import requests
from bs4 import BeautifulSoup
from collections import deque
```

```python
%%time 
q = deque()
questions_list = []
main_url = 'http://questions.chgk.info'
q.append(f'/all.html')
while q:
    url = q.pop()
    page = requests.get(f'{main_url}{url}')
    soup = BeautifulSoup(page.content, 'html.parser')
    for link in soup.find_all('a'):
        href = link.get('href')
        if href.endswith('-q.html'):
            questions_list.append(href)
        elif href.startswith('/files/') and not href.endswith('-a.html'):
            q.append(href)
```

```python
len(questions_list)
```

```python
%%time
all_q = []
for l in questions_list:
    page = requests.get(f'{main_url}{l}')
    soup = BeautifulSoup(page.content, 'html5lib')
    all_q += [' '.join(el.get_text().split('\n')[1:]) for el in soup.find_all('a') 
              if el.get_text().startswith('Вопрос')]
```

```python
with open('questions_list.txt', 'w') as f:
    f.write("\n".join(questions_list))
```
